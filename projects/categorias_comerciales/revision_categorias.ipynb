```python
from db_config_data_fact import engine_data_fact

# Verificar que la conexión está establecida
print("Conexión con engine_data_fact establecida")
```

```python
# Descargar la tabla categorizacion_establecimientos_2026_01
df_categorias = pd.read_sql("SELECT * FROM categorizacion_establecimientos_2026_01", engine_data_fact)

print(f"Tabla descargada exitosamente")
print(f"Dimensiones: {df_categorias.shape}")
print(f"\nPrimeras filas:")
print(df_categorias.head())
```

```python
# Valores únicos de la columna categoria
categorias_unicas = df_categorias['categoria'].unique()

print(f"Categorías únicas: {len(categorias_unicas)}")
print("\nListado de categorías:")
for i, cat in enumerate(categorias_unicas, 1):
    print(f"{i}. {cat}")
```

```python
# Leer el archivo categorias_.xlsx
ruta_excel = r'c:\Users\ruben.freire\OneDrive - PUBLIPROMUEVE S.A\Escritorio\Proyectos\CategorizarEstablecimientos\categorias_.xlsx'
df_excel = pd.read_excel(ruta_excel)

print("Archivo categorias_.xlsx cargado exitosamente")
print(f"Dimensiones: {df_excel.shape}")
print(f"\nColumnas: {df_excel.columns.tolist()}")
print(f"\nPrimeras filas:")
print(df_excel.head())
```

```python
# Categorías únicas del archivo Excel
# Detectar la columna que contiene categorías (usualmente contiene 'categ', 'categoria', etc.)
columna_categoria = [col for col in df_excel.columns if 'categ' in col.lower()][0] if any('categ' in col.lower() for col in df_excel.columns) else df_excel.columns[0]

categorias_unicas_excel = df_excel[columna_categoria].unique()

print(f"Categorías únicas del Excel (columna '{columna_categoria}'): {len(categorias_unicas_excel)}")
print("\nListado de categorías del Excel:")
for i, cat in enumerate(categorias_unicas_excel, 1):
    print(f"{i}. {cat}")
```

```python
# Comparar categorías únicas de ambas fuentes
bd_set = set(categorias_unicas)
excel_set = set(categorias_unicas_excel)

# Categorías en BD, Excel o ambas
solo_bd = bd_set - excel_set
solo_excel = excel_set - bd_set
ambas = bd_set & excel_set

# Crear dataframe comparativo
comparacion_data = []

# Agregar categorías comunes
for cat in sorted(ambas):
    comparacion_data.append({'Categoría': cat, 'En BD': '✓', 'En Excel': '✓'})

# Agregar categorías solo en BD
for cat in sorted(solo_bd):
    comparacion_data.append({'Categoría': cat, 'En BD': '✓', 'En Excel': ''})

# Agregar categorías solo en Excel
for cat in sorted(solo_excel):
    comparacion_data.append({'Categoría': cat, 'En BD': '', 'En Excel': '✓'})

df_comparacion = pd.DataFrame(comparacion_data)

print(f"Resumen de comparación:")
print(f"  - Total en BD: {len(bd_set)}")
print(f"  - Total en Excel: {len(excel_set)}")
print(f"  - En ambas: {len(ambas)}")
print(f"  - Solo en BD: {len(solo_bd)}")
print(f"  - Solo en Excel: {len(solo_excel)}")
print(f"\nDataframe comparativo:")
print(df_comparacion)
```
